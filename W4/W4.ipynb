```python
import pandas as pd
data = pd.read_csv('adult.data.csv', header=None)
data.head()
```

```python
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
              'marital_status', 'occupation', 'relationship', 'race', 'gender',
              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
data.head(10)
```

```python
data.columns
data['income'][32557]
```

```python
#姓別和收入關係
gender_income = data[data['income']== ' >50K'].groupby('gender')['income'].count()
print(gender_income)
```

```python
#高收入者若是男性，其教育程度是否較高?
gender_income_edu = data[(data['income'] == ' >50K') & (data['education_num'] == 14)].groupby('gender')['income'].count()
print(gender_income_edu)
```

```python
#工時和高收入關係
import numpy as np

# 篩選出收入大於50K的數據
high_income_data = data[data['income'] == ' >50K']

# 計算工作小時數的平均值、標準差和中位數
average_hours_per_week = high_income_data['hours_per_week'].mean()
std_dev_hours_per_week = high_income_data['hours_per_week'].std()
median_hours_per_week = high_income_data['hours_per_week'].median()

print("平均工作小時數:", average_hours_per_week)
print("工作小時數標準差:", std_dev_hours_per_week)
print("工作小時數中位數:", median_hours_per_week)
```
