```python
import pandas as pd
data = pd.read_csv('adult.data.csv', header=None)
data.head()
```

```python
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
              'marital_status', 'occupation', 'relationship', 'race', 'gender',
              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
data.head(10)
```

```python
data.columns
data['income'][32557]
```

```python
#姓別和收入關係
gender_income = data[data['income']== ' >50K'].groupby('gender')['income'].count()
print(gender_income)
```

```python
#高收入者若是男性，其教育程度是否較高?
gender_income_edu = data[(data['income'] == ' >50K') & (data['education_num'] == 14)].groupby('gender')['income'].count()
print(gender_income_edu)
```

```python
#工時和高收入關係
import numpy as np

# 篩選出收入大於50K的數據
high_income_data = data[data['income'] == ' >50K']

# 計算工作小時數的平均值、標準差和中位數
average_hours_per_week = high_income_data['hours_per_week'].mean()
std_dev_hours_per_week = high_income_data['hours_per_week'].std()
median_hours_per_week = high_income_data['hours_per_week'].median()

print("平均工作小時數:", average_hours_per_week)
print("工作小時數標準差:", std_dev_hours_per_week)
print("工作小時數中位數:", median_hours_per_week)
```

```python
import plotly.express as px

# 过滤数据以获取高收入男性和高收入女性的教育程度信息
filtered_data = data[data['income'] == ' >50K']

# 使用 Plotly Express 创建小提琴图，将 'gender' 作为颜色分组
fig = px.violin(filtered_data, y='education_num', color='gender', title='Education Level of High-Income Individuals by Gender')
fig.show()
```

```python
import plotly.express as px

# 过滤数据以获取高收入男性和高收入女性的教育程度信息
filtered_data = data[data['income'] == ' >50K']

# 使用 Plotly Express 创建小提琴图，将 'gender' 作为颜色分组
fig = px.violin(filtered_data, y='education_num', color='gender', title='Education Level of High-Income Individuals by Gender')

# 计算中位数
median_education_male = filtered_data[filtered_data['gender'] == ' Male']['education_num'].median()
median_education_female = filtered_data[filtered_data['gender'] == ' Female']['education_num'].median()

# 在图表上添加中位数标记
fig.add_hline(y=median_education_male, line_dash="dash", line_color="blue", name="Median Education (Male)")
fig.add_hline(y=median_education_female, line_dash="dash", line_color="red", name="Median Education (Female)")

fig.show()
```
